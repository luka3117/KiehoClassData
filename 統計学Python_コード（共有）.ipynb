```
# List 3-3 パッケージstatisticsの算術平均関数を使う
import statistics      # <-- statisticsパッケージをimport
height = [168.3, 179.2, 165.8, 170.5, 188.2, 
          174.6, 162.8, 175.5, 178.1, 177.1]
average = statistics.mean(height)   # importしたstatisticsパッケージの中のmeanを使う
print(average)

```

```
# List 3-4 パッケージstatisticsの算術平均関数を使
from statistics import mean      # <-- statisticsパッケージからすべてをimport
height = [168.3, 179.2, 165.8, 170.5, 188.2, 
          174.6, 162.8, 175.5, 178.1, 177.1]
average = mean(height)   # <--importした中にmeanも含まれている
print(average)
```

```
# List 3-5 パッケージstatisticsのメジアン・モード関数を使う
from statistics import median, mode      # <-- statisticsパッケージからmedianとmodeをimport
print(median(
    [168.3, 179.2, 165.8, 170.5, 188.2, 
    174.6, 162.8, 175.5, 178.1, 177.1]
    )
)
print(mode([19, 21, 19, 20, 22, 19, 20, 21, 20, 20,]))
```

# 幾何平均

```
# List 3-6 パッケージscipy.statsの幾何平均関数を使う
from scipy.stats import gmean      # <-- scipy.statsパッケージからmedianとmodeをimport
print(gmean([1, 2, 4, 8, 16])) 
# 出力結果は　4.0
```

```
#-*- coding: utf-8 -*-
# List 3-7 幾何平均を自前で計算する
import numpy as np
x = np.array([1, 2, 4, 8, 16])
print( np.power(x.prod(), 1/len(x)) )
```

```
# 出力結果は　4.0# -*- coding: utf-8 -*-
# List 3-8 トリム平均の計算
from scipy import stats
x = [168.3, 179.2, 165.8, 170.5, 188.2, 
    174.6, 162.8, 175.5, 178.1, 177.1]
print(stats.trim_mean(x, 0.1))  # xに対して両側から0.1ずつトリムして平均
```

```
# 図全体を出力（表示）する
# # List 4-1  １７歳身長統計を正規分布と比較する
from scipy.stats import norm
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import japanize_matplotlib # <- 日本語を表示

# filename="/content/drive/MyDrive/山内pythonCode/ch4/h28_学校保健統計_身長分布.csv"
filename = "./h28_学校保健統計_身長分布.csv" <- text book
# filename = "https://raw.githubusercontent.com/luka3117/ClassData/master/KyotoEDU/h28_%E5%AD%A6%E6%A0%A1%E4%BF%9D%E5%81%A5%E7%B5%B1%E8%A8%88_%E8%BA%AB%E9%95%B7%E5%88%86%E5%B8%83.csv" # <- github


df = pd.read_csv(filename)
```

```
df2 = df.iloc[61:110,[0,13]].astype(float)  #17歳のみ切り出し、
df2 = df2.rename(columns={'Unnamed: 0':'height', 'Unnamed: 13':'permil'}) # 欄名を変更

u = [norm.pdf(x=i, loc=170.7, scale=5.81)*1000.0 for i in df2['height']] # 正規分布関数値を生成
df2['norm'] = u  # データフレームに追加

print(df2)
print(sum(df2.iloc[:,1]), sum(df2.iloc[:,2]))

# import japanize_matplotlib # <- 日本語を表示

ax = df2.plot.scatter(x='height',y='permil', color='black', marker='x', label='身長統計')
df2.plot(x='height',y='norm', color='black', kind='line', ax=ax, 
         label='N(170.7, 5.81)')  #重ねて描くためにax=axとする
plt.grid()     #グリット線を引いてくれる
plt.xlabel('身長')     #x軸のラベル
plt.ylabel('パーミル‰')     #y軸のラベル
plt.legend(loc="upper left")   #凡例を左上に表示
plt.title('17歳男子の身長統計と正規分布')
plt.show()
```
